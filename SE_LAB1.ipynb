```python
#program 1
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time):
  # Hard coded coeficients for quadratic equation
  a = 0.1
  b = -1
  c = 30
  #quadratic equation representing temparature change
  temperature = a*(time**2) + b*time + c
  return temperature

def main():
    #time value for 0 to 10 example
    time_values = np.linspace(0,10,50)
    #calculate temp using hardcode coefficients
    temperature_hardcoded = quadratic_model(time_values)
    #plot results
    plt.plot(time_values, temperature_hardcoded, label = "Hard-coded coefficients")
    plt.xlabel("Time")
    plt.ylabel("temperature")
    plt.legend
    plt.title("Weather modeling with quadratic equation (Hard-coded coefficients)")
    plt.show()

if __name__ == "__main__":
      main()
```

```python
#program 5
import numpy as np
import matplotlib.pyplot as plt

def read_coefficients(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

        # Initialize lists to store coefficients
        a_values = []
        b_values = []
        c_values = []

        # Read coefficients from each line
        for line in lines:
            coeffs = line.strip().split()
            if len(coeffs) == 3:
                a_values.append(float(coeffs[0]))
                b_values.append(float(coeffs[1]))
                c_values.append(float(coeffs[2]))

        return a_values, b_values, c_values

def quadratic_model(a, b, c, time):
    # Calculate temperature for multiple sets of coefficients
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    filename = '/content/Monica2.txt'
    a_values, b_values, c_values = read_coefficients(filename)

    time_values = np.linspace(0, 10, 50)

    # Plotting multiple curves
    for a, b, c in zip(a_values, b_values, c_values):
        temp_file = quadratic_model(a, b, c, time_values)
        plt.plot(time_values, temp_file, label=f"Coefficients: ({a}, {b}, {c})")

    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.legend()
    plt.title("Weather Modeling with Quadratic Equations")
    plt.show()

if __name__ == "__main__":
    main()
```
